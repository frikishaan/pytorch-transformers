# Translation with Transformer Model
This notebook demonstrates how to use the trained transformer model for translation.

```python
# Import required libraries
import torch
from inference import generate_response, CONFIG, DEVICE, SRC_SP, TGT_SP, MODEL

# Set the model to evaluation mode
MODEL.eval()

def translate(text: str, max_length: int = 100) -> str:
    """
    Translate the input text from source to target language.
    
    Args:
        text: Input text to translate
        max_length: Maximum length of the generated translation
    Returns:
        Translated text
    """
    with torch.no_grad():
        translation = generate_response(text, max_length)
    return translation
```

```python
# Example usage
input_text = "the sky color is blue"
translation = translate(input_text)
print(f"Input: {input_text}")
print(f"Translation: {translation[1]}")
```
